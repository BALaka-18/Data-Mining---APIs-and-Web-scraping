```python
rank = []
grade = []
channel_name = []
uploads = []
subsc = []
views = []


for i in container[4:15]:
    
    rank.append(i.find('div',attrs={'style':'float: left; width: 50px; color:#888;'}).text.strip())
    try:
        grade.append(i.find('div',attrs={'style':'float: left; width: 70px; font-size: 1.1em;'}).find('span',attrs={'style':'font-weight: bold; color:#00bee7;'}).text.strip())
    except:
        grade.append('NaN')
    channel_name.append(i.find('a').text.strip())
    try:
        uploads.append(i.find('div',attrs={'style':'float: left; width: 80px;'}).find('span',attrs={'style':'color:#555;'}).text.strip())
    except:
        uploads.append('NaN')
    pop_list = i.find_all('div',attrs={'style':'float: left; width: 150px;'})
    subsc.append(pop_list[0].text.strip())
    try:
        views.append(pop_list[1].find('span').text.strip())
    
    except:
        views.append('NaN')
        
for i in container[15:65]:
    rank.append(i.find('div',attrs={'style':'float: left; width: 50px; color:#888;'}).text.strip())
    try:
        grade.append(i.find('div',attrs={'style':'float: left; width: 70px; font-size: 1.1em;'}).find('span',attrs={'style':'font-weight: bold; color:#69ce10;'}).text.strip())
    except:
        grade.append('NaN')
    channel_name.append(i.find('a').text.strip())
    try:
        uploads.append(i.find('div',attrs={'style':'float: left; width: 80px;'}).find('span',attrs={'style':'color:#555;'}).text.strip())
    except:
        uploads.append('NaN')
    pop_list = i.find_all('div',attrs={'style':'float: left; width: 150px;'})
    subsc.append(pop_list[0].text.strip())
    try:
        views.append(pop_list[1].find('span').text.strip())
    
    except:
        views.append('NaN')
        
for i in container[65:]:
    rank.append(i.find('div',attrs={'style':'float: left; width: 50px; color:#888;'}).text.strip())
    try:
        grade.append(i.find('div',attrs={'style':'float: left; width: 70px; font-size: 1.1em;'}).find('span',attrs={'style':'font-weight: bold; color:#5ac10e;'}).text.strip())
    except:
        grade.append('NaN')
    channel_name.append(i.find('a').text.strip())
    try:
        uploads.append(i.find('div',attrs={'style':'float: left; width: 80px;'}).find('span',attrs={'style':'color:#555;'}).text.strip())
    except:
        uploads.append('NaN')
    pop_list = i.find_all('div',attrs={'style':'float: left; width: 150px;'})
    subsc.append(pop_list[0].text.strip())
    try:
        views.append(pop_list[1].find('span').text.strip())
    
    except:
        views.append('NaN')

d = {'Rank':rank,'Grade':grade,'Ch_name':channel_name,'Uploads':uploads,'Subscriptions':subsc,'Views':views}
final1 = pd.DataFrame(d)
```

```python
final1
```

```python
final1.to_csv('top_500.csv',index=False)
```

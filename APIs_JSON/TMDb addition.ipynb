```python
resp = requests.get('https://api.themoviedb.org/3/genre/movie/list?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US').json()
idx = []
genre_name = []
final = pd.DataFrame()
    
for i in resp['genres']:
    idx.append(i['id'])
    genre_name.append(i['name'])
        
d = {'id':idx, 'genre':genre_name}
df = pd.DataFrame(d)
final = final.append(df,ignore_index=True)
```
